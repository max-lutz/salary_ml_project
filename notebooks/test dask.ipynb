```python
import sys
import dask.dataframe as dd
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
df = pd.DataFrame(np.random.randn(100000,20))
print(sys.getsizeof(df))
```

```python
da = dd.from_pandas(df, npartitions=4)
print(sys.getsizeof(da))
```

```python
for i in tqdm(range(100)):
    df = pd.DataFrame(np.random.randn(100000,200))
    df.to_hdf('test.h5',key='jobs', append=True, mode='a', format='table')
```

```python
df = pd.read_hdf('test.h5')
print(sys.getsizeof(df))
del df
```
